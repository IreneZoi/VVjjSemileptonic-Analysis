```python
import ROOT
import os
from pyroot_cms_scripts import CMS_style, CMS_text
```

```python
set_variable_defaults = """
hist_filename = "signal_loose_W_mu.root"
variable = "n2_sdb1_tau2tau1"
title_x = ""
title_y = ""
tag = "dataM"
text = "dataM"
draw_opt = "colz1"
fill_style = ""
fill_color = ""
plots_dir = "plots"
"""
```

```python
plot_mc_data = """
print("-----------------------")
print(hist_filename)
print("-----------------------")

# Force CMS_style
# ==============
ROOT.TGaxis().SetMaxDigits(5)
CMS_style.SetLabelSize(0.04, "XYZ")
CMS_style.SetCanvasDefW(650)
CMS_style.SetPadRightMargin(0.12)
CMS_style.cd()
ROOT.gROOT.ForceStyle()

hist_file = ROOT.TFile.Open(hist_filename)

if tag == "data":
    h = hist_file.Get(f"dataM_{variable}")
    h.Add(hist_file.Get(f"dataE_{variable}"))
else:
    h = hist_file.Get(f"{tag}_{variable}")
    
if title_x != "":
    h.GetXaxis().SetTitle(title_x)
if title_y != "":
    h.GetYaxis().SetTitle(title_y)

latex = ROOT.TLatex(0.2, 0.85, text)
latex.SetNDC()
latex.SetTextFont(42)

if fill_style != "":
    h.SetFillStyle(fill_style)
if fill_color != "":
    h.SetFillColor(fill_color)
h.SetMarkerStyle(20)
h.SetMarkerSize(0.5)

canvas = ROOT.TCanvas()

h.Draw(draw_opt)

canvas.Draw()
latex.Draw()
"""
```

```python
exec(set_variable_defaults)
hist_filename = "loose_no_V_mass_cut_W.root"
variable = "mass_V_n2_sdb1"
title_x = "m_{V}"
title_y = "N_{2}"

os.makedirs(f"{plots_dir}", exist_ok=True)

tag = "data"
text = "data"
exec(plot_mc_data)
canvas.SaveAs(f"{plots_dir}/{tag}_{variable}_{hist_filename.split('/')[-1].split('.')[0]}.pdf")

tag = "WJets"
text = "WJets"
exec(plot_mc_data)
canvas.SaveAs(f"{plots_dir}/{tag}_{variable}_{hist_filename.split('/')[-1].split('.')[0]}.pdf")

tag = "VBS_EWK"
text = "VBS_EWK"
exec(plot_mc_data)
canvas.SaveAs(f"{plots_dir}/{tag}_{variable}_{hist_filename.split('/')[-1].split('.')[0]}.pdf")

tag = "VBS_QCD"
text = "VBS_QCD"
exec(plot_mc_data)
canvas.SaveAs(f"{plots_dir}/{tag}_{variable}_{hist_filename.split('/')[-1].split('.')[0]}.pdf")

tag = "Top"
text = "Top"
exec(plot_mc_data)
canvas.SaveAs(f"{plots_dir}/{tag}_{variable}_{hist_filename.split('/')[-1].split('.')[0]}.pdf")
```

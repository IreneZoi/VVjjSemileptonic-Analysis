```python
import ROOT
import os
from pyroot_cms_scripts import CMS_style, CMS_text
```

```python
set_variable_defaults = """
hist_filename = "from_ram_ntuples/signal_loose_W_mu.root"
variable = "n2_sdb1_tau2tau1"
title_x = ""
title_y = ""
tag = "dataM"
text = "dataM"
plots_dir = "plots"
"""
```

```python
plot_mc_data = """
print("-----------------------")
print("")
print(hist_filename)
print("")
print("-----------------------")

# Force CMS_style
# ==============
ROOT.TGaxis().SetMaxDigits(5)
CMS_style.SetLabelSize(0.04, "XYZ")
CMS_style.SetCanvasDefW(650)
CMS_style.SetPadRightMargin(0.12)
CMS_style.cd()
ROOT.gROOT.ForceStyle()

hist_file = ROOT.TFile.Open(hist_filename)

if tag == "data":
    h = hist_file.Get(f"dataM_{variable}") + hist_file.Get(f"dataE_{variable}")
else:
    h = hist_file.Get(f"{tag}_{variable}")
    
if title_x != "":
    h.GetXaxis().SetTitle(title_x)
if title_y != "":
    h.GetYaxis().SetTitle(title_y)

latex = ROOT.TLatex(0.2, 0.85, text)
latex.SetNDC()
latex.SetTextFont(42)

canvas = ROOT.TCanvas()

h.Draw("colz1")

os.makedirs(f"{plots_dir}", exist_ok=True)
canvas.Draw()
latex.Draw()
canvas.SaveAs(f"{plots_dir}/{tag}_{variable}_{hist_filename.split('/')[-1].split('.')[0]}.pdf")
"""
```

```python
exec(set_variable_defaults)
variable = "n2_sdb1_tau2tau1"
title_x = "N_{2}"
title_y = "#tau_{21}"
tag = "dataM"
text = "dataM"
exec(plot_mc_data)
```
